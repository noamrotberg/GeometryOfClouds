# 3d-Example for module

This is an example how our module can be used to compute connected components of data on a regular triangle tiling with data in several hight-layers.

Import the module for connected components ... (vollständer Titel hier einfügen):

```python
import module_3dccl_vers7 as ccl
ccl.path_input(path = '/work/bb1018/b380906/getting_started/', 
               gridfile = 'icon-grid_nawdex_78w40e23n80n_R80000m.nc',
               datafile = 'nawdexnwp-80km-mis-0001_2016092200_3dcloud_DOM01_ML_0020.nc')
```

Define radius and start_triangle:

```python
radius = 30
start_triangle = 5736
```

Note the following radius choices in our grid for this start_triangle = 5736:
* radius = 102 to obtain the whole grid

* radius = 56 for maximal area not touching the boundaries

start_triangle can be chosen between 0 and 7919

Determine the cube coordinates:

```python
cube_coordinates = ccl.make_cube_coordinates (start_triangle, radius, print_progress = True)
```

It is possible to save the cube coordinates in a local ".npy"-file and load it later on. In this way, the cube coordinates for the same grid, start_triangle and radius must not be computed again for different field data.

```python
ccl.save_cube_coordinates(cube_coordinates, start_triangle, radius, "cube_coordinates")
```

```python
cube_coordinates = ccl.load_cube_coordinates("cube_coordinates-start_5736-radius_30.npy")
```

There are two types for connectivity on triangles: vertex and edge connectivity. Our module supports both of them. 

We start by considering vertex connectivity and later move on to edge connectivity.

### Vertex connectivity

Calculate the connected components and write them into lists. Highlight the components in plot of the grid by setting "make_plot = True". (Note that this needs latitude and longitude coordinates for the triangles, see paper!)

```python
components_list_vertex = ccl.make_connected_3d_components(cube_coordinates,
                                                          radius,
                                                          connectivity = 'vertex', 
                                                          make_plot = True, 
                                                          save_plot = True, 
                                                          save_name = 'connected_components_rad6.pdf', 
                                                          plot_title = 'connected components on icon')
```

### Edge connectivity 

Calculate the connected components and write them into lists. Highlight the components in plot of the grid by setting "make_plot = True". (Note that this needs latitude and longitude coordinates for the triangles, see paper!)

```python
components_list_edge = ccl.make_connected_3d_components(cube_coordinates,
                                                        radius,
                                                        connectivity = 'edge', 
                                                        make_plot = True, 
                                                        save_plot = False, 
                                                        save_name = 'connected_components.pdf', 
                                                        plot_title = 'connected components on icon')
```
